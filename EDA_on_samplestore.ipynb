```python
plt.figure(figsize=(17,8))
plt.bar('Sub-Category','Category',data=data,color='g')
plt.xlabel('Sub Category',size=15)
plt.ylablel('Category',size=15)
plt.xticks(rotation=30)
plt.show()
```

```python
plt.figure(figsize=(15,11))
data['Sub-Category'].value_counts().plot.pie(autopct='%1.1f%%')
plt.show()
```

```python
data.groupby(['Category','Sub-Category'])['Profit','Sales'].sum().plot.bar(figsize=(10,8))
plt.title('Total Profit and sales per Sub Category')
plt.legend(loc='upper center')
plt.show()
```

```python
data.groupby('State')['Profit','Sales'].sum().plot.bar(figsize=(15,10))
plt.show()
```

```python
plt.figure(figsize=(17,10))
sns.countplot(x='Sub-Category',hue='Region',data=data)
plt.show()
```

```python
plt.figure(figsize=(17,10))
sns.countplot(x='Region',hue='Ship Mode',data=data)
plt.show()
```

```python
plt.figure(figsize=(10,8))
data.groupby('Ship Mode').Profit.sum().plot.pie(autopct='%1.1f%%',explode=(0,0.1,0,0.1),shadow=True)
plt.title('Profit percent in each shipping mode')
plt.show()
```

```python
plt.figure(figsize=(15,10))
sortedTop20 =data.sort_values(['Profit'],ascending=False).head(20)
t_tick=list(sortedTop20['State'])
fig,((ax1,ax2)) =plt.subplots(1,2)
ax1.set_title("Top 20 state having high profits")
sortedTop20.plot('State','Profit',kind='bar',ax=ax1)
sortedTop20.plot(x='State',y='Profit',kind='line',figsize=(15,8),ax=ax2)
plt.xticks(list(range(len(t_ticks))),t_tick,fontsize=12,rotation=90)
plt.show()
```

```python
sortedbtm20=data.sort_values(['Profit'],ascending=False).tail(20)
b_tick=list(sortedbtm20['State'])
sortedbtm20.plot(x='State',y='Profit',kind='line',figsize=(15,8))
plt.xticks(list(range(len(b_tick))),b-tick,fontsize=12,rotation=90)
plt.title('top 20 states having low profits')
plt.show()
```
